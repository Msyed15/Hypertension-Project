```python
import pandas as pd

def readingData(firstYear, lastYear) :
  data = []
  while (firstYear <= lastYear) :
    year = str(firstYear)
    data.append(pd.read_csv('https://raw.githubusercontent.com/CDavisBelmont/HypertensionTN/main/TN_Local_Data_for_Better_Health_AllYears_release_Tableau_final.xlsx%20-%20' + year + '_release.csv'))
    firstYear = firstYear + 1
  return data
```

```python
data = readingData(2016,2020)
data[0].head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

corr = data[0].corr()  # calculate the correlation matrix of the entire data frame
fig, ax = plt.subplots()  # create a new figure object and subplot
cax = ax.matshow(corr, cmap='coolwarm', vmin=-1, vmax=1)  # plot the correlation matrix
fig.colorbar(cax)  # add a colorbar to the plot
ticks = np.arange(0,len(data[0].columns),1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(data[0].columns, rotation=90)  # rotate x-axis labels by 90 degrees
ax.set_yticklabels(data[0].columns)
plt.show()  # display the plot
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select certain columns from the DataFrame
selected_columns = ['Annual Checkup', 'Dental Visit', 'Physical Inactivity']
selected_data = data[0][selected_columns]

# Create a correlation matrix for the selected columns
corr_matrix = selected_data.corr()

# Plot the correlation matrix as a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```
